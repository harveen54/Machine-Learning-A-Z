```python
#importing required libraries
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.mlab as mlab
import math
```

# Effect on Linear Function on Gaussian

```python
#creating a normal distribution
mu = 0
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x,mlab.normpdf(x, mu, sigma))
plt.title("Gaussian Function")
plt.show()
```

```python
y= 2*x + 2
plt.plot(range(len(y)),y)
plt.title("Linear function")
plt.show()
```

```python
#observing the final output after passing a Gaussian into a Linear function

plt.plot(y,mlab.normpdf(y, mu, sigma))


plt.title("Output Gaussian")
plt.show()
```

# Effect of Non Linear function on Gaussian

```python
#creating a normal distribution
mu = 0
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x,mlab.normpdf(x, mu, sigma))
plt.title("Gaussian Function")
plt.show()
```

```python
#creating a non linear function y
y= np.arctan(x)
plt.plot(range(len(y)),y)
plt.title("Non Linear function")
plt.show()
```

```python
plt.plot(y,mlab.normpdf(y, mu, sigma))
plt.title("Output Non Gaussian")
plt.show()
```

```python

mu, sigma = 0, 0.1

s = np.random.normal(mu, sigma, 1000)
count, bins, ignored = plt.hist(s, 30, normed=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *np.exp( - (bins - mu)**2 / (2 * sigma**2) ),linewidth=2, color='r')
plt.show()
```

```python
y= np.arctan(s)
from sklearn import preprocessing
y_scaled = preprocessing.scale(y)
mu, sigma = 0, 0.1
count, bins, ignored = plt.hist(y_scaled, 30, normed=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *np.exp( - (bins - mu)**2 / (2 * sigma**2) ),linewidth=2, color='r')
plt.show()
```
